```python
import pandas as pd
import numpy as np
```

```python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
```

```python
#計算每個不同種類 animal 的 age 的平均數
for i in np.unique(df.animal):  print(str(i)+':',np.mean(df.age[df.animal==i]))
for i in set(df.animal):  print(str(i)+':',np.mean(df.age[df.animal==i]))
print(df.groupby(['animal']).mean(),'\n')
#將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
print(df.sort_values(by=['age']).sort_values(by=['visits'], ascending=False),'\n')

df1=df.sort_values(by=['age'])#same as above
print(df1.sort_values(by=['visits'], ascending=False),'\n')
#將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
df["priority"]=df["priority"].str.replace("yes","True")
df["priority"]=df["priority"].str.replace("no","False")
print(df)
```

```python
#2.一個包含三個欄位的 DataFrame，將每個數字減去
df = pd.DataFrame(np.random.random(size=(5, 3)))
df
```

```python
#1) 該欄位的平均數
print(df-df.mean()) #Broadcast
f = lambda x: x - x.mean() #lambda
print(df.apply(f))
#2) 該筆資料平均數
print(df-df.stack().mean())
print(df-df.values.mean())
```

```python
#承上題，請問：
#1) 哪一行的資料總合最小
#2) 哪一欄位的資料總合最小

print(df.mean(axis=0).argmin())
print(df.mean(axis=1).argmin())
```

```python
import pandas as pd
score_df = pd.DataFrame([[1,56,66,70], 
              [2,90,45,34],
              [3,45,32,55],
              [4,70,77,89],
              [5,56,80,70],
              [6,60,54,55],
              [7,45,70,79],
              [8,34,77,76],
              [9,25,87,60],
              [10,88,40,43]],columns=['student_id','math_score','english_score','chinese_score'])
score_df = score_df.set_index('student_id')
score_df
```

```python
#6號學生(student_id=6)3科平均分數為何?
print(score_df.loc[6].mean(),'\n') #equal to score_df.iloc[5].mean()
#6號學生3科平均分數是否有贏過班上一半的同學?
print("number 6 student's mean score is higher only ",100*sum(score_df.mean(axis=1)<score_df.loc[6].mean())/len(score_df),'% of classmates','\n')
#由於班上同學成績不好，所以學校統一加分，加分方式為開根號乘以十，請問6號同學3科成績分別是?
score_df_new = score_df.apply(lambda x : x**(0.5)*10)
print(score_df_new.loc[6],'\n')
#承上題，加分後各科班平均變多少?
print(score_df_new.mean())
```
