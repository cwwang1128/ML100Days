```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

boston = pd.read_csv(r'C:/Users\user\Downloads\boston.csv')
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
boston.plot.scatter(x = 'NOX', y = 'DIS')
df_corr = boston[['NOX', 'DIS']].corr(method = 'pearson')
print(df_corr)#負相關
```

2.    給定散點圖顏色

```python

# 設定要畫的 x, y 數據 list....
x = np.arange(0,180)
y = np.cos( x*np.pi/180.0 )
```

```python

plt.plot(x,y)
plt.xlim(0, 200)
plt.ylim(-1, 1.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python

# 畫出完整的 sin 圖形
x = np.arange( 0, 2*np.pi, 0.1 )
y_sin = np.sin(x)

plt.plot( x, y_sin )
plt.title('sin')
plt.show()
#plt.savefig("sin.png",dpi=300,format="png")
```

```python

# 畫出完整的 sin 圖形
y_cos = np.cos(x)

plt.plot( x, y_cos )
plt.title('cos')
plt.show()
#plt.savefig("cos.png",dpi=300,format="png")
```

```python
#給定散點圖顏色
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y,X)
plt.scatter(X,Y, s=75, c=T, alpha=0.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
